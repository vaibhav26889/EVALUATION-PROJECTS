# LINEAR REGRRESION MODEL

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
max_r_score=0
for r_state in range(40,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20, random_state=r_state)
    linr=LinearRegression()
    linr.fit(x_train,y_train)
    predlinr=linr.predict(x_test)
    r2score=r2_score(y_test,predlinr)
    if(r2score>max_r_score):
        max_r_score=r2score
        final_r_state=r_state
print('')
print('The max r2score corresponing to the',final_r_state,'is',max_r_score)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20, random_state=99)
```

```python
linr.fit(x_train,y_train)
linr.score(x_train,y_train)
```

```python
predlinr=linr.predict(x_test)
```

```python
print('r2score=',r2_score(y_test,predlinr))
print(mean_absolute_error(y_test,predlinr))
print('RMSE=',np.sqrt(mean_squared_error(y_test,predlinr)))
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

NOW, TO INCREASE THE SCORE WE WILL APPLY ENSEMPLE TECHNIQUES

```python
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
rf=RandomForestRegressor(n_estimators=100, random_state=99)
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
```

```python
predrf=rf.predict(x_test)
print('r2Score =',r2_score(y_test,predrf))
```

After using RandomForestRegressor technique we can see that score is increased to 0.9751 and R2score is 0.8435

```python
ad=AdaBoostRegressor(random_state=99)
ad.fit(x_train,y_train)
ad.score(x_train,y_train)
```

```python
predad=ad.predict(x_test)
print('r2score=', r2_score(y_test,predad))
```

After using AdaBoostRegressor we can see that normal score is 0.6550 & r2score is 0.6621

```python
gb=GradientBoostingRegressor()
gb.fit(x_train,y_train)
gb.score(x_train,y_train)
```

```python
predgb=gb.predict(x_test)
print('r2score=',r2_score(y_test,predgb))
```

After using GradientBosting method we can see that score is 0.8394 and r2score is 0.7390

# CLASSIFICATION METHOD

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
df_1
```

```python
x1=df_1.iloc[:,0:10]
x1
```

```python
y1=df_1.iloc[:,-1]
y1
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.20,random_state=99)
```

```python
logr=LogisticRegression()
```

```python
logr.fit(x1_train,y1_train)
logr.score(x1_train,y1_train)
```

```python
predlogr=logr.predict(x1_test)
```

```python
print('Accuracy Score=',accuracy_score(y1_test,predlogr))
print('Confusion Matrix:',confusion_matrix(y1_test,predlogr))
print(classification_report(y1_test,predlogr))
```

```python
model=[MultinomialNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier()]
for m in model:
    m.fit(x1_train,y1_train)
    m.score(x1_train,y1_train)
    predm=m.predict(x1_test)
    print('Accuracy-Score of',m, 'is')
    print(accuracy_score(y1_test,predm))
    print(confusion_matrix(y1_test,predm))
    print(classification_report(y1_test,predm))
    print('\n')
```
