```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
max_r_score=0
for r_state in range(40,105):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20, random_state=r_state)
    linr=LinearRegression()
    linr.fit(x_train,y_train)
    predlinr=linr.predict(x_test)
    r2score=r2_score(y_test,predlinr)
    if(r2score>max_r_score):
        max_r_score=r2score
        final_r_state=r_state
print("")
print('The max r2score corresponding to the',final_r_state,'is',max_r_score)
```

Here we can see that maximum value of r2_score is 85.5% on random_state=64

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=64)
```

```python
linr.fit(x_train,y_train)
linr.score(x_train,y_train)
```

```python
predlinr=linr.predict(x_test)
```

```python
print('r2Score =',r2_score(y_test,predlinr))
print(mean_absolute_error(y_test,predlinr))
print('RMSE =',np.sqrt(mean_squared_error(y_test,predlinr)))
```

# WE CAN TRY SOME ENSEMBLE TECHNIQUES

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor
```

```python
rf=RandomForestRegressor(n_estimators=200,random_state=64)
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
```

```python
predrf=rf.predict(x_test)
print('r2Score =',r2_score(y_test,predrf))
```

AFTER USING RANDOM FOREST REGRESSSOR TECHNIQUE WE CAN SEE THAT VALUE OF R2 SCORE IS 0.7683 WHICH IS LESS THEN THE ACTUAL SCORE

```python
ad=AdaBoostRegressor(random_state=64)
ad.fit(x_train,y_train)
ad.score(x_train,y_train)
```

```python
predad=ad.predict(x_test)
print('r2score=', r2_score(y_test,predad))
```

WE CAN SEE THAT WHILE USING ADA BOOST REGRESSOR NORMAL SCORE IS IMPROVED TO 0.9940 AND R2SCORE IS 0.6550

```python
gb=GradientBoostingRegressor()
gb.fit(x_train,y_train)
gb.score(x_train,y_train)
```

```python
predgb=gb.predict(x_test)
print('r2score=',r2_score(y_test,predgb))
```

WE CAN SEE THAT AFTER USING GRADIENT BOOSTING TECHNIQUE NORMAL SCORE IS IMPROVED TO 0.9999 AND R2 SCORE IS .06177
