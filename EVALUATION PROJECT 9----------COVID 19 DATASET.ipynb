# NOW LETS TRAIN OUR MODEL AND PUT OUR TARGET VARIABLE(Y) AS DEATHS

```python
x1=df_new.iloc[:,0:2]
x1.head()
```

```python
x2=df_new.iloc[:,3:13]
x2.head()
```

```python
x=pd.concat([x1,x2],axis=1)
x.head()
```

```python
x.shape
```

```python
y=df_new.iloc[:,2]
y.head()
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
max_r_score=0
for r_state in range(10,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20, random_state=r_state)
    linr=LinearRegression()
    linr.fit(x_train,y_train)
    predlinr=linr.predict(x_test)
    r2score=r2_score(y_test,predlinr)
    if(r2score>max_r_score):
        max_r_score=r2score
        final_r_state=r_state
print('')
print('The max r2score corresponing to the',final_r_state,'is',max_r_score)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20, random_state=90)
```

```python
linr.fit(x_train,y_train)
linr.score(x_train,y_train)
```

```python
predlinr=linr.predict(x_test)
```

```python
print('R2 Score=',r2_score(y_test,predlinr))
print(mean_absolute_error(y_test,predlinr))
print('RMSE=',np.sqrt(mean_squared_error(y_test,predlinr)))
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
predrf=rf.predict(x_test)
print('R2Score',r2_score(y_test,predrf))
```

```python
ad=AdaBoostRegressor(random_state=90)
ad.fit(x_train,y_train)
ad.score(x_train,y_train)
```

```python
predad=ad.predict(x_test)
print('R2Score',r2_score(y_test,predad))
```

GRADIENT BOOSTING
